```python
class DataClean():
    def __init__(self) -> None:
        pass

    def remove_pairs(self, data, key):
        try:
            del data[key]
        except Exception as e:
            pass
        
    def cleanner(self, data, keyss=['start', 'end', 'token_start', 'token_end']):
        try:
            clean_data = []
            for d in data:
                dt_data = d
                dt_tokens = []
                for token in d['tokens']:
                    dt_token = {}
                    for k in token.keys():
                        if k not in keyss:
                            dt_token[k] = token[k]
                    dt_tokens.append(dt_token)
                dt_data['tokens'] = dt_tokens
                clean_data.append(dt_data)
            
        except Exception as e:
               pass


    def token_cleanner(self, data, data_keys=[], token_keys=[]):
        for df in data:
            for key in data_keys:
                self.remove_pairs(df, key)

        self.cleanner(data, keyss=token_keys)
```

```python
cleanner = DataClean()
```

```python
cleanner.token_cleanner(train_data, data_keys=['relations'], token_keys=['start', 'end', 'token_start', 'token_end'])
```
